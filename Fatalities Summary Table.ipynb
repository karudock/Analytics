### Add column for region 
- using functions and then .apply

```python
def region(row):
    if row['STATE'] in Northeast:
        return 'Northeast'
    if row['STATE'] in West:
        return 'West'
    if row['STATE'] in South:
        return 'South'
    if row['STATE'] in Midwest:
        return 'Midwest'
    if row['STATE'] =="United States":
        return 'United States'
    if row['STATE'] =="Northeast":
        return 'Northeast'
    if row['STATE'] =="West":
        return 'West'
    if row['STATE'] =="South":
        return 'South'
    if row['STATE'] =="Midwest":
        return 'Midwest'

us['Region'] = us.apply(lambda row: region(row), axis=1)

us.head(100)
#us.to_csv(r"us_more.csv")
```

```python
# curious about the number of people involved in accidents

g= tblAcc.groupby(['STATE', 'YEAR'], as_index=False).agg({"PERSONS":"sum", "PEDS":"sum"})
g
g['Total_Number_People']=g.PERSONS + g.PEDS
g=g.rename(columns={"PERSONS":"TotNum_Vehicle", "PEDS":"TotNum_Pedestrians"})
g
f= tblAcc.groupby(['STATE', 'YEAR'], as_index=False).agg({"FATALS":"sum"})
f.rename(columns={'Fatals':'Number_of_Fatalities'})


more=pd.merge(g,f,on=['STATE', 'YEAR'], how='inner')
```

```python
# add region in here too bc why not I love more work

Northeast = ['Maine', 'New Hampshire', 'Vermont', 'Massachusetts', 'Rhode Island', 'Connecticut', 'New York', 'New Jersey', 'Pennsylvania']
Midwest = ['Ohio', 'Michigan', 'Indiana', 'Wisconsin', 'Illinois', 'Minnesota', 'Iowa', 'Missouri', 'North Dakota', 'South Dakota', 'Nebraska','Kansas']
South = ['District of Columbia','Delaware', 'Maryland', 'Virginia', 'West Virginia', 'Kentucky', 'North Carolina', 'South Carolina', 'Tennessee', 'Georgia', 'Florida', 'Alabama', 'Mississippi', 'Arkansas', 'Louisiana', 'Texas', 'Oklahoma']
West = ['Montana', 'Idaho', 'Wyoming', 'Colorado', 'New Mexico', 'Arizona', 'Utah', 'Nevada', 'California', 'Oregon', 'Washington', 'Alaska', 'Hawaii']


def region(row):
    if row['STATE'] in Northeast:
        return 'Northeast'
    if row['STATE'] in West:
        return 'West'
    if row['STATE'] in South:
        return 'South'
    if row['STATE'] in Midwest:
        return 'Midwest'

more['Region'] = more.apply(lambda row: region(row), axis=1)
more.head(100)
```

```python
# merge with population

spop=pd.read_csv(r"US_pop.csv")
us3=pd.merge(more,spop,on=['STATE','YEAR'],how='left')
us3['Standardized_Fatalities'] = (us3['FATALS'] / (us3['Population']/100000))
us3['Standardized_Pedestrians'] = (us3['TotNum_Pedestrians'] / (us3['Population']/100000))
us3['Standardized_People'] = (us3['Total_Number_People'] / (us3['Population']/100000))

us3.sort_values(by=['STATE', 'YEAR'])

us3['Pct_Change_Pop_State'] = us3.groupby(['STATE'])['Population'].apply(lambda x: x.div(x.iloc[0]).subtract(1).mul(100))

us3['Pct_Change_Fatalities_State'] = us3.groupby(['STATE'])['Standardized_Fatalities'].apply(lambda x: x.div(x.iloc[0]).subtract(1).mul(100))

us3['Pct_Change_People_State'] = us3.groupby(['STATE'])['Standardized_People'].apply(lambda x: x.div(x.iloc[0]).subtract(1).mul(100))

us3['Pct_Change_Pedestrians_State'] = us3.groupby(['STATE'])['Standardized_Pedestrians'].apply(lambda x: x.div(x.iloc[0]).subtract(1).mul(100))

us3.to_csv("trythis_US_people.csv")
```
